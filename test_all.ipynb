```python
import scipy.io
import pylab as plt
import numpy as np
import os, glob, pickle
from read_rsr import read_rsr
# read the vegetation spectra from DLR database
spectra_fn = 'spectrals/Spectres.sav'
spectra = scipy.io.readsav(spectra_fn)
int_spec = {}
#m = len(spectra.spectres)
for i in glob.glob('RSRs/*RSR.txt'):
    sensor = read_rsr(i)
    name = i.split('_RSR')[0].split('/')[1]
    int_spec[name] = get_int_refs(sensor, swl=spectra.wl, specs = spectra.spectres)
```

```python
sensor = 'terra'
plt.figure(figsize=(15,3))
plt.imshow( int_spec[sensor]['int_refs'], interpolation='none' )
print int_spec[sensor]['cwls']
plt.show()
```

```python
mappings, trains = mapping(int_spec, filename = 'test_mapping.txt')
```

```python
# define here sensorA and sensorB
# sensorA is the one to map to using band observations from sensorB
sensorA = 'sentinel2'
sensorB = 'aqua'

#print sensorA, sensorB

# for 27 randomly selected spectra from the DLR database...
nrows = 6 # for image display
ncols = 3 # for image display
fig = plt.figure( figsize = (15, 20))
num_subplots = nrows * ncols
integrated_spectra = int_spec
test_idxs = np.delete(np.arange(len(spect)), trains)


for i, spectrum_number in enumerate(test_idxs[:num_subplots]):
    cwlsA = integrated_spectra[sensorA]['cwls']
    m = len(cwlsA) # m number of bands in sensorA
    bandsA = range(m)
    spectrum = np.zeros((m, 4), dtype=float)
    for bandA in bandsA:
        cwlA = integrated_spectra[sensorA]['cwls'][bandA]
        #y = integrated_spectra[sensorA]['int_refs'][:,bandA].T

        integreflB = integrated_spectra[sensorB]['int_refs'].T

        b_nums = mappings[sensorA].keys()
        b = b_nums[bandA]
        bandsB = mappings[sensorA][b][sensorB]['bandsB']
        var_y_given_X = mappings[sensorA][b][sensorB]['var_y_given_X']
        beta_hat = mappings[sensorA][b][sensorB]['beta_hat']

        X_ = integreflB[:, bandsB]
        X = np.matrix( np.hstack(( np.ones( np.shape(X_)[0] ).reshape(( np.shape(X_)[0], 1)), X_ )) ) # create design matrix

        y_pred, var_y = evaluate(X, X, beta_hat, var_y_given_X)

        yp = float(y_pred[spectrum_number,0]) # y_pred is a column vector, we still index in full 2 dimensions...
        sigma = np.sqrt(float(var_y[spectrum_number]))
        t = 1.96
        y_lower = yp - (t * sigma)
        y_upper = yp + (t * sigma)

        spectrum[bandA, :] = y_lower, yp, y_upper, sigma

    # plot in order of centre wavelengths
    e = np.argsort(cwlsA)
    ax = fig.add_subplot(nrows, ncols, i+1)

    #ax.set_ylim((0, 0.6))
    ax.locator_params(axis='y',nbins=4)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(10)
    ax.set_xlabel("wavelength (nm)", fontsize=10)


    if i+1 == 1 or i+1 == 4 or i+1 == 7 or i+1 == 10 or i+1 == 13 or i+1 == 16:
        for tick in ax.yaxis.get_major_ticks():
                tick.label.set_fontsize(10)
        ax.set_ylabel("reflectance", fontsize=10)
    else:
        ax.set_yticklabels(())

    plt.plot(cwlsA[e], integrated_spectra[sensorA]['int_refs'].T[spectrum_number, e], 'o', alpha=0.3)
    plt.errorbar(cwlsA[e], spectrum[e,1], yerr=1.96*spectrum[e,3], fmt='.', color='black')
    # in case you may want to interpolate between bands...
    plt.plot(cwlsA[e], spectrum[e,1], 'k-.')
    plt.fill_between(cwlsA[e], spectrum[e,0], spectrum[e,2], facecolor='grey', alpha = 0.3)
    plt.title(' %s and %s spec# %s' % (sensorA, sensorB,spectrum_number))
    plt.locator_params(nbins=5)

plt.tight_layout()
```
